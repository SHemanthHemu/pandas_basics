## Data Selection Techniques 

```python
import pandas as pd
```

```python
employee_df = pd.read_csv(filepath_or_buffer=r"C:\Users\heman\Desktop\chat gpt\data\input\realistic_employee_data.csv")
```

```python
# Selection by Label 
employee_df.loc[0]
```

```python
# Select rows and specific columns by label 
employee_df.loc[0,'Department']
```

```python
# Select a subset of rows and columns by labels 
employee_df.loc[0:6,["Department","Salary"]]
```

```python
# Select a specific cell by label 
employee_df.loc[5,"Salary"]
```

```python
# Select rows from 0 to 5 and all columns 
employee_df.loc[0:5]
```

```python
# Select rows with specific condition and columns 
employee_df.loc[0:6,'Name':"Salary"]
```

```python
# Selection by Position 
```

```python
# Select the first row using position 
employee_df.iloc[0]
```

```python
# Select a specific row and column by position 
employee_df.iloc[1,3]
```

```python
# Select a range of rows and columns by position 
employee_df.iloc[0:6,4]
```

```python
# Select all rows for a specific column by position 
employee_df.iloc[:,2]
```

```python
#  Slicing with .iloc[] 
employee_df.iloc[0:8]
```

```python
# Select rows from position 0 to 5 and all columns 
employee_df.iloc[0:5,:]
```

```python
# Select rows and specific columns by position 
employee_df.iloc[0:6,3:5]
```

```python
# Boolean Indexing 
```

```python
# Basic Boolean Indexing 
```

```python
# Filter rows where 'Age' is greater than 30 
employee_df[employee_df["Age"]>30]
```

```python
# Filter rows where 'Salary' is less than 50000 
employee_df[employee_df["Salary"]<80000]
```

```python
# Combine multiple conditions using '&' (and) and '|' (or) 
employee_df[(employee_df["Age"]>30)&(employee_df["Salary"]<80000)]
```

```python
# Use 'isin' to filter rows where a column contains one of a set of values 
employee_df[employee_df["Department"].isin(["Finance","marketing"])]
```

```python
# Rows where 'Age' is greater than 30 and 'Salary' is greater than 40000 
employee_df[(employee_df["Age"]>30)& (employee_df["Salary"]>40000)]
```

```python
# Rows where 'Category' is either 'Tech' or 'Finance', and 'Salary' is greater than 50000 
employee_df[(employee_df["Department"].isin(["Finaace","Engineering"]))& (employee_df["Salary"]>50000)]
```

```python
# Using .query() for Conditional Selection 
```

```python
# Query with a condition 
employee_df.query('Age > 50')
```

```python
# Using multiple conditions in the query 
employee_df.query('Age > 50 & Salary > 50000')
```

```python
# Query with column names that have spaces 
```

```python
# Advanced Querying with .query() 
# Use variables inside query 

min_age = 30
max_Salary = 50000

employee_df.query('Age > @min_age & Salary < @max_Salary')
```

```python
# Selecting Data Based on Index 
```

```python
# Select data by row index (label-based) 
employee_df.loc[5]
```

```python
# Select data by position (position-based) 
employee_df.iloc[6]
```

```python
# Select rows with index labels from 'A' to 'C' (inclusive) 
employee_df.loc[5:8]
```

```python
# Select rows from index position 0 to 5 
employee_df.iloc[0:6]
```

```python
# Selecting Columns 
```

```python
# Select a single column by label 
employee_df['Department']
```

```python
# Select multiple columns by labels 
employee_df[['Department','Age']]
```

```python
# Select all columns except for specific ones 
employee_df.drop(['Gender'],axis=1)
```

```python
# Selecting Columns by Position 
```

```python
# Select a single column by position (e.g., column 2) 
employee_df.iloc[:,2]
```

```python
# Select multiple columns by position (e.g., columns 1 and 2) 
employee_df.iloc[:,1:5]
```

```python
# Selecting Rows and Columns Simultaneously 
```

```python
# Select rows with label 'A' to 'C' and columns 'Age' and 'Salary' 
employee_df.loc[1:3,['Age',"Salary"]]
```

```python
# Select specific row and column 
employee_df.loc[3,'Department']
```

```python
# Select rows from position 0 to 5 and columns from position 1 to 3 
employee_df.iloc[0:6,1:3]
```

```python
# Selection Using .at[] and .iat[] 
```

```python
# Using .at[] for Label-based Access 
# Access a single value by label (row label and column label) 
employee_df.at[1,'Salary']
```

```python
# Access a single value by position (row position and column position) 
employee_df.iat[2,3]
```
